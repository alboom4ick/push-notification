```python
import requests
import json
import time

url = "http://10.36.40.28:8888/api/main/create-notification/"
params = {"time": "2025-09-14T10:00:00Z"}
for i in range(56, 61):
    response = requests.post(url+str(i), params=params)
    key = json.loads(response.text)
    print(f" ----- Key is: {key}")
    while True:
        result_url = "http://10.36.40.28:8888/api/main/notification/data-keys/"
        result = requests.get(result_url+str(key))
        result = json.loads(result.text)
        print(result)
        try:
            push_notification = result['result']["message"]
            best_product = result['best_product']
            with open("output.csv", "a", encoding="utf-8") as f:
                f.write(f"{i},{best_product},{json.dumps(push_notification, ensure_ascii=False)}\n")
                print(f"Completed: {i}, {best_product}")
            break
        except KeyError:
            print("Notification is not ready yet. Please try again later.")
            time.sleep(1)
            continue
```
